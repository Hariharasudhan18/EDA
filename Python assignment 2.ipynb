```python
n=5;
for i in range(n):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```

### Write a Python program to reverse a word after accepting the input from the user.

```python
word = input("Input word: ")
print("output:",word[::-1])
```
